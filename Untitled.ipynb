```python
# Removing outliers from the data set
import numpy as np
```

```python
from scipy import stats
z_score = np.abs(stats.zscore(df[['LIMIT_BAL']]))
```

```python
outlier_condition = (z_score > 3)
df_cleaned = df[~outlier_condition.any(axis=1)]
```

```python
df_cleaned
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numeric_cols = ['LIMIT_BAL', 'AGE', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 
                'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns=['default.payment.next.month'])
```

```python
X
```

```python
y = df['default.payment.next.month']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 42)
```

```python
X_train.head()
```

```python
y_train
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression 
model = LogisticRegression()
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score

accuracy = accuracy_score(y_test, y_pred)
```

```python
accuracy
```

```python
roc_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
print(f'ROC AUC: {roc_auc}')
```
